# 1. Setting Up

```python
path1, path2 = "./images/desk_W_orig.png", "./images/desk_C_orig.png"
img_w, img_c = read_and_resize(path1), read_and_resize(path2)
fig, axes = plt.subplots(1, 2)
axes[0].imshow(img_w, cmap="gray")
axes[0].set_title("world frame")
axes[1].imshow(img_c, cmap="gray")
axes[1].set_title("camera frame")

for axis in axes:
    axis.axis("off")
plt.show()

print(f"{img_w.shape}, {img_c.shape}")
```

```python
H, W = img_w.shape
K = np.array([[100, 0, W // 2],
             [0, 100, H // 2],
             [0, 0, 1]])
# K = np.eye(3)
# np.linalg.inv(K)
```

```python
pts_w_img = np.array([[23, 326],
                 [158, 727],
                 [569, 524],
                 [500, 218],
                 [506, 270],
                 [135, 372],
                 [238, 656],
                 [565, 498],
                 [513, 420],
                 [489, 365],
                 [424, 434],
                 [317, 597]])

pts_w_plane_img = np.array([[506, 270],
                        [135, 372],
                        [238, 656],
                        [565, 498],
                        [513, 420]])

pts_c_img = np.array([[287, 123],
                 [5, 390],
                 [376, 718],
                 [541, 462],
                 [531, 475],
                 [321, 209],
                 [145, 408],
                 [414, 666],
                 [447, 557],
                 [472, 502],
                 [399, 478],
                 [234, 458]])

pts_c_plane_img = np.array([[531, 475],
                        [321, 209],
                        [145, 408],
                        [414, 666],
                        [447, 557]])
```

```python
camera_coord = []
for pts in [pts_w_img, pts_w_plane_img, pts_c_img, pts_c_plane_img]:
    pts_h = np.empty((pts.shape[0], 3))
    for i in range(pts.shape[0]):
        pts_h[i, :] = to_homogenous(pts[i, :])
#     print(f"pts_h: {pts_h}")
    pts_camera_coord = (np.linalg.inv(K) @ pts_h.T).T
    
    camera_coord.append(pts_camera_coord[:, :-1] / pts_camera_coord[:, -1:])
    # camera_coord.append(pts_camera_coord / pts_camera_coord[:, -1:])

pts_w, pts_w_plane, pts_c, pts_c_plane = camera_coord

# for pts in [pts_w, pts_w_plane, pts_c, pts_c_plane]:
#     print(pts)
#     print("-" * 50)
```
